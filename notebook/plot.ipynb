Fig 6a

PDF Plot of Conformations

```python
from scipy.interpolate import make_interp_spline
nbins = 50
def make_pdf(r):
    count,xx = np.histogram(r,nbins)
    p = count / np.sum(count)
    x = [(xx[i]+xx[i+1])/2 for i in range(nbins)]
    x_s = np.linspace(np.min(x),np.max(x),2000)
    p_s = make_interp_spline(x,p)(x_s)
    return x_s,p_s

data = np.load("../mono/2smallerlooprmsd_ndp_mono.npz",allow_pickle=True)
r1 = np.hstack(data["rmsd1s"])
data = np.load("../mono/2smallerlooprmsd_ndp_tyk_mono.npz",allow_pickle=True)
r2 = np.hstack(data["rmsd1s"])
data = np.load("../oct/2smallerlooprmsd_ndp_oct.npz",allow_pickle=True)
r3 = np.hstack(data["rmsd1s"])
data = np.load("../oct/2smallerlooprmsd_ndp_tyk_oct.npz",allow_pickle=True)
r4 = np.hstack(data["rmsd1s"])
plt.figure(figsize=(12,6),dpi=500)
sns.kdeplot(r1,label="Monomer-Allosteric Tylosin",color="purple",alpha=0.6,lw=4)
print(np.sum(r1<3)/len(r1),np.sum(r1>11)/len(r1))
sns.kdeplot(r2,label="Monomer+Allosteric Tylosin",color="blue",alpha=0.6,lw=4)
print(np.sum(r2<3)/len(r2),np.sum(r2>11)/len(r2))
sns.kdeplot(r3,label="Octamer-Allosteric Tylosin",color="orange",alpha=0.6,lw=4)
print(np.sum(r3<3)/len(r3),np.sum(r3>6)/len(r3))
sns.kdeplot(r4,label="Octamer+Allosteric Tylosin",color="red",alpha=0.6,lw=4)
print(np.sum(r4<3)/len(r4),np.sum(r4>6)/len(r4))
plt.xlim(0,15)
plt.ylim(0,1.4)
plt.ylabel("Density", y=0.5/1.4,fontdict={
    "family":"Arial","size":14.4,"weight":"bold"})
plt.xlabel("RMSD of 232-237 (Å)",fontdict={
    "family":"Arial","size":14.4,"weight":"bold"})
plt.xticks(fontproperties="Arial",fontsize=12)
plt.yticks(np.linspace(0,1,11),fontproperties="Arial",fontsize=12)
ax = plt.gca()
ax.spines["top"].set_linewidth(0)
ax.spines["right"].set_linewidth(0)
plt.legend(frameon=False,prop={"family" : "Arial","size":14.4},)
plt.show()
```
